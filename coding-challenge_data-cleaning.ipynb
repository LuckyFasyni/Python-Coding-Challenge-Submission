#### Instruksi 1 : Load data yang sudah disediakan (telco_customer_churn.csv) kedalam dataFrame

```python
import pandas as pd

#load data yang mempunyai format csv
df = pd.read_csv('telco_customer_churn.csv', sep=',')

#menampilkan lima baris teratas dari dataset
df.head()
```

#### Instruksi 2 : Tunjukkan tipe data untuk semua kolom yang ada!

```python
df.dtypes
```

#### Instruksi 3 : Ubah nama kolom!
* CustomerID menjadi CustID
* Semua Kolom menjadi lowercase (huruf kecil)

```python
#mengubah nama customerID menjadi custID
df = df.rename(columns = {'customerID' : 'custID'})

# mengubah semua nama kolom menjadi huruf kecil (lower case) 
df.columns = df.columns.str.lower()

df.head()
```

#### Instruksi 4 : Hilangkan kata (automatic) dari value kolom PaymentMethod,
pastikan juga tidak ada spasi yang tersisa di awal/akhir! (hint: gunakan string functions)

```python
#mengecek value dari kolom PaymentMethod yang memiliki "(automatic)"
df['paymentmethod'].value_counts()
```

```python
#Menghilangkan (automatic) dari value
df['paymentmethod'] = df['paymentmethod'].str.replace(' \(automatic\)', '')

#Menghilangkan spasi yang ada di awal dan di akhir value
df['paymentmethod'] = df['paymentmethod'].str.strip()

df.head()
```

## Data Preprocessing

#### Instruksi 5 : One-hot encoding untuk kolom phoneservice dengan prefix â€˜serviceâ€™

```python
#membuat one hot encoding dengan prefix 'service'
df = pd.get_dummies(df, columns=['phoneservice'], prefix='service')
df.head()
```

#### Instruksi 6 : Normalization untuk kolom monthlycharges

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()

df['monthlycharges_normalized'] = min_max_scaler.fit_transform(df[['monthlycharges']])
df.head()
```

```python
#mengecek hasil normalisasi
df['monthlycharges_normalized'].describe()
```

range value dengan value min = 0 dan value max = 1 setelah dilakukan normalization

#### Instruksi 7 : Standardization untuk kolom totalcharges

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()

#Tipe data totalcharges sebelumnya adalah object, maka kita harus mengubahnya menjadi numerik/float
#menghilangkan spasi yang mungkin ada di value
df['totalcharges'] = df['totalcharges'].str.replace(' ', '')
#mengubah tipe data ke numerik
df['totalcharges'] = pd.to_numeric(df['totalcharges'])
#menangani value NaN yang muncul. Jika ada NaN, maka akan diganti dengan mean dari totalcharges 
if df['totalcharges'].isnull().sum() != 0:
    df['totalcharges'] = df['totalcharges'].fillna(value=df['totalcharges'].mean())

#standardization process
df['totalcharges_standardized'] = std_scaler.fit_transform(df[['totalcharges']])
df.head()
```

```python
#mengecek hasil standardization
df['totalcharges_standardized'].describe()
```

nilai mean sudah bernilai 0 dan standar deviasi sudah bernilai 1 setelah dilakukan standardization
